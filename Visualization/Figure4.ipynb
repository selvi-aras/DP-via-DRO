```python
# %matplotlib notebook
%matplotlib inline
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.patches as mpatches
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [10, 6]
from mpl_toolkits import mplot3d
```

```python
prob_dists = pd.read_csv("/.../Visualise/multi/mid.csv", header=None, sep=',')
distributions = prob_dists.to_numpy()
beta = 0.50
limits = 25.0
X = np.arange(-limits, limits + 10**(-5), beta)
k = prob_dists.shape[0]
```

```python
counts = [np.concatenate([np.repeat((X[i] + X[i+1])/2, round((10**6)*prob_dists.iloc[j][i])) for \
                           i in range(prob_dists.iloc[j].shape[0])]) for j in range(k)]
```

```python
d_to_plot = [0,4,8] #1, 5, 9-th distributions
```

```python
# plt.style.use('seaborn-deep')
plt.style.use('default')
plt.xlim(-15,15)
plt.ylim(0,0.35)
colors = ["blue", "red","forestgreen"]
for t in range(3):
    plt.hist(counts[d_to_plot[t]], X, label=["measure#" + str(d_to_plot[t]+1)], density = True, color = colors[t], alpha = 0.2, rwidth =1.0);
plt.legend(loc='upper right')
#do not show the legend
# plt.legend().set_visible(False)
plt.tight_layout()
# plt.show()
plt.title(r"$\epsilon =$ "+str(0.6) + r", $\delta = $" + str(0.2), fontsize=16)
plt.savefig("save_mid.pdf", format='pdf',bbox_inches='tight')
```
